# handling null values

```python
ws=pd.read_csv('Windspeed.csv')
```

```python
ws
```

```python
ws.info()
```

```python
ws.isnull().sum()
```

```python
ws.isnull().mean()
```

```python
#filling a common value to all missing values
ws.fillna(0)
```

```python
#adding missing values in individual columns
ws.fillna({
        "Temperature":0,
        "Windspeed":5,
        "Status":"sunny"
})
```

```python
ws
```

```python
#FORWARD FILL
ws.fillna(method="ffill")
```

```python
#backward fill
ws.fillna(method='bfill')
```

```python
#limit
ws.fillna(method="ffill", limit=1)
```

```python
#filling with pandas objects
ws.fillna(ws.mean())
```

```python
#filling for specific range
ws.fillna(ws.mean()["Temperature":"Windspeed"])
```

```python
#interpolate
```

```python
ws.interpolate()
```

```python
ws.interpolate(limit=1,limit_direction="backward")
```

```python
ws.interpolate(limit_direction="both",limit=1,limit_area="inside")
```

```python
ws.interpolate(limit_direction="both",limit=1,limit_area="outside")
```
