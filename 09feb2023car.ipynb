```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
car=pd.read_csv('Car_sales.csv')
car
```

```python
car.dropna(inplace=True)
car.info()
```

```python
car.Latest_Launch.unique()
```

```python
car['Latest_Launch']= pd.to_datetime(car.Latest_Launch)
car.info()
```

```python
car['year']=pd.DatetimeIndex(car.Latest_Launch).year
car
```

```python
car['month']=pd.DatetimeIndex(car.Latest_Launch).month
```

```python
car['day']=pd.DatetimeIndex(car.Latest_Launch).day
```

```python
car
```

```python
car.info()
```

```python
car.Fuel_efficiency=car.Fuel_efficiency.astype(int)
car.info()
```

```python
car.month.replace({
    1:"JANUARY",
    2:"FEBRUARY",
    3:"MARCH",
    4:"APRIL",
    5:"MAY",
    6:"JUNE",
    7:"JULY",
    8:"AUGUST",
    9:"SEPTEMBER",
    10:"OCTOBER",
    11:"NOVEMBER",
    12:"DECEMBER"
    
},
inplace=True)
```

```python
car["lemonth"]=le.fit_transform(car.month)
```
